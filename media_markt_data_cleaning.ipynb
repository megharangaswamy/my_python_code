```python
# Import modules
import pandas as pd

# Set ipython's max row display
pd.set_option('display.max_row', 1000)

# Set iPython's max column width to 50
pd.set_option('display.max_columns', 50)
```

```python
#read a csv file
df= pd.read_csv('F:/web_data_int/project/media_markt/media_markt _1.csv', sep=',')
df.tail()
```

```python
#droping unwanted columns
list(df)
df = df.drop(['web-scraper-order', 'web-scraper-start-url','pagination','pagination-href','Processor_Number'], axis=1)
df.head()
```

```python
#extrcting data from title

#adding required columns
df.insert(0, 'MM_id', 'M_')
df.insert(1, 'Brand', 'brand_value')
df.insert(2, 'laptop_name', 'laptop_name_value')
df.insert(3, 'display_info', 'display_info')
df.insert(4, 'processor_info', 'aaaa')
df.insert(5, 'RAM', 'RAM_value')
df.insert(6, 'HDD', 'HDD_value')
df.insert(7, 'grapic_Card', 'grapic_card_value')
df.insert(8, 'colour', 'colour_value')

#extrcting data from name column
for i in range(0,len(df)):
    df['MM_id'][i] = "M_"+str(i) # MM specific id
    title = df['laptop'][i] #loptop column contain the tile value
    df['Brand'][i] = title.split(',')[0].split(' ')[0]
    df['laptop_name'][i] = title.split(',')[0]
    df['display_info'][i] = title.split(',')[1]
    df['processor_info'][i] = title.split(',')[2]
    if(df['Brand'][i] == 'APPLE'):
        df['RAM'][i] = 'null'
        df['HDD'][i] = title.split(',')[3]
        df['grapic_Card'][i] = title.split(',')[4]
        df['colour'][i] = title.split(',')[5]
    else:
        df['RAM'][i] = title.split(',')[3]
        df['HDD'][i] = title.split(',')[4]
        df['grapic_Card'][i] = title.split(',')[5]
        df['colour'][i] = title.split(',')[6]
    df['MM_Item_number'][i] = int(df['MM_Item_number'][i]) # converting float to int

```
